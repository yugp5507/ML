```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import OneHotEncoder
```

```python
badminton = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/tennis_anyone.csv")
badminton = badminton.rename(columns={'Play-Tennis': 'play-badminton'})
badminton
```

```python
badminton.head()
```

```python
badminton.shape
```

```python

badminton_encoded = pd.get_dummies(badminton, dtype = int)
badminton_encoded
```

```python
X = badminton_encoded.drop(columns = ['play-badminton_Yes'])
y = badminton_encoded['play-badminton_Yes']
```

```python
classifier = GaussianNB()
classifier.fit(X,y)
```

```python
predictions = classifier.predict(X)
```

```python
Confusion_matrix = pd.crosstab(predictions, y, rownames=['predicted'], colnames=['actual'])
print(Confusion_matrix)
```

```python
for idx, feature in enumerate(X.columns):
    print(f"\n Table for feature '{feature}':")
    if idx < len(classifier.theta_):
        print(f"Mean: {classifier.theta_[idx]}")
        print(f"Variance: {classifier.var_smoothing}")
    else:
        print("No data available for this feature")
```
